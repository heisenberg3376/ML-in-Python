# EXPLORATORY DATA ANALYSIS

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('titanic.csv')
df.head()
```

## 1.Categorical Data

### a. Countplot

```python
sns.countplot(x=df['Embarked'])
```

### b. Pie chart

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

## 2. Numerical Data

### a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=10)
```

### b. distplot

probability density function<br>
if you select a random passenger, then the prob that his age is x is y

```python
sns.distplot(df['Age'])
```

### c. boxplot

i) min<br>
ii) Q1<br> 25% values are lesser than this value<br>
iii) median<br>
iv) Q2<br>75% values are lesser than this value<br>
v) max<br>

```python
sns.boxplot(df['Age'])
```

```python
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
```

```python
tips.head()
```

# Bivariate/Multivariate analysis

## 1. Scatterplot (numeric - numeric)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

## 2. barplot

```python
titanic.head()
#titanic['parch'].value_counts()
```

```python
sns.barplot(x=titanic['pclass'],y=titanic['age'],hue=titanic['sex'])
```

## 3. Boxplot

```python
sns.boxplot(x=titanic['sex'],y=titanic['age'],hue=titanic['survived'])
```

## 4. distplot

```python
sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)
sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)
```

## e. Heatmap (cat- cat)

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
print(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
(titanic.groupby('sex').mean()['survived']*100).plot(kind='bar')
print(titanic.groupby('sex').mean()['survived']*100)
# % of people survived in each pclass
```

## 6. Cluster map

```python
sns.clustermap(pd.crosstab(titanic['parch'],titanic['survived']))
```

## g. lineplot

x axis value must be time based

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
new = flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
